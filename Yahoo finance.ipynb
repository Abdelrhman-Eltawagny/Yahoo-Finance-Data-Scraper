```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
from selenium import webdriver
```

```python
page = requests.get('https://finance.yahoo.com/most-active?count=25&offset=0')
content = page.content
soup = BeautifulSoup(content, features='html.parser')
```

```python
symbol = []
name_of_company = []
price = []
change = []
percentage_change = []
Volume = []
avg_volume = []
Market_Cap = []
PE_Ratio = []
```

```python
offset = 0
while True : 
    page = requests.get(f'https://finance.yahoo.com/most-active?count=25&offset={offset}')
    content = page.content
    soup = BeautifulSoup(content, features='html.parser')

    companies = soup.find('tbody').find_all('tr')
    
    for i in range(len(companies)):
        symbol.append(companies[i].contents[0].text)
        name_of_company.append(companies[i].contents[1].text)
        price.append(companies[i].contents[2].text)
        change.append(companies[i].contents[3].text)
        percentage_change.append(companies[i].contents[4].text)
        Volume.append(companies[i].contents[5].text.replace('M' , '').strip())
        avg_volume.append(companies[i].contents[6].text.replace('M' , '').strip())
        Market_Cap.append(companies[i].contents[7].text)
        PE_Ratio.append(companies[i].contents[8].text)
        
    print(f'{offset} : Done')
    offset += 25
    if offset == 225 :
        print('done all')
        break
```

```python
stock_price = pd.DataFrame({'Symbol' : symbol , 'Company Name' : name_of_company , 'Stock Price' : price , 'Change' : change ,
                            'Percentage Change' : percentage_change , 'Volume(M)' : Volume , 'Average Volume Month(M)' : avg_volume ,
                            'Market Cap' : Market_Cap , 'PE Ratio(TTM)' : PE_Ratio})
```
